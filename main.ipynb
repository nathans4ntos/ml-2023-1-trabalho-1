```python
import pandas as pd
import numpy as np
```

```python
# carrega o arquivo xlsx em um dataframe do pandas
dataset = pd.read_excel('dataset.xlsx')
dataset
```

```python
# detecta dados faltantes
column_null = dataset.isnull().sum().sort_values(ascending=False)
column_not_null = dataset.notnull().sum().sort_values(ascending=False)
column_null_percent = (column_null/(column_null + column_not_null))*100
dataset_null_percent = pd.concat({'Nulos': column_null, 'Não nulos': column_not_null, 'Porcentagem': column_null_percent}, axis=1)
dataset_null_percent[dataset_null_percent['Porcentagem']>=90]
```

```python
# cria uma lista das colunas nulas ou quase nulas a serem deletadas
columns_null = dataset_null_percent[dataset_null_percent['Porcentagem']>=90].index.to_list()
columns_null
```

```python
# dropa as colunas
dataset.drop(columns=columns_null, inplace=True)
dataset
```

```python
# Substitui os valores em string para number para facilitar a substituição de nulos restantes pela média
dataset.replace('not_detected', 0, inplace=True)
dataset.replace('detected', 1, inplace=True)
dataset.replace('negative', 0, inplace=True)
dataset.replace('positive', 1, inplace=True)
dataset.replace('absent', 0, inplace=True)
dataset.replace('present', 1, inplace=True)

# Faz a substituição dos valores nulos pela média da coluna
for i in dataset.columns[dataset.isnull().any(axis=0)]:   
    dataset[i].fillna(dataset[i].mean(),inplace=True)
    
dataset
```

```python
# iterar pelas colunas do DataFrame
for col in dataset.columns:
    # pula a coluna de ID e as colunas de booleanos para calcular os outliers
    if col in ["Patient ID", 
              "SARS-Cov-2 exam result", 
              "Patient addmited to semi-intensive unit (1=yes, 0=no)", 
              "Patient addmited to intensive care unit (1=yes, 0=no)", 
              "Patient addmited to regular ward (1=yes, 0=no)",
              "Mean platelet volume "]:
        continue
    # calcular o limite superior e inferior para detecção de outliers
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_sup = Q3 + 1.5 * IQR
    lim_inf = Q1 - 1.5 * IQR

    # identificar os outliers
    outliers = dataset[(dataset[col] < lim_inf) | (dataset[col] > lim_sup)]

    # imprimir o número de outliers e seus valores máximo e mínimo
    print(f"Coluna: {col}")
    print(f"Número de outliers: {len(outliers)}")
    print(f"Máximo outlier: {outliers[col].max()}")
    print(f"Mínimo outlier: {outliers[col].min()}")
    print("\n")

dataset.to_excel('dataset_novo.xlsx', index= False)
```
